## Complex Functions

$$z = |z|e^{i(\varphi + 2k\pi)} \in \mathbb{Z}$$

```python
# Preamble

%config InlineBackend.figure_formats = ['retina']
%config InlineBackend.figure_formats = ['svg']

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.rcParams['axes.prop_cycle'] = plt.cycler(color=["olivedrab", "steelblue", "goldenrod", "firebrick"]) 
plt.rcParams['figure.figsize'] = [8, 5]
plt.rcParams['figure.constrained_layout.use'] = True
plt.rcParams['legend.frameon'] = False
plt.rcParams["xtick.minor.visible"] = True
plt.rcParams["ytick.minor.visible"] = True
```

#### Square Root

$$\sqrt{z\,} = |z|^{1/\kern+1pt2}e^{i(\varphi / \kern+1pt 2 + k\pi)} = |z|^{1/\kern+1pt2}
[\cos(\varphi / \kern+1pt 2 + k\pi) + i \sin(\varphi / \kern+1pt 2 + k\pi)]$$

#### Natural Logarithm

$$\log(z) = \log(|z|) + \log\left(e^{i(\varphi + 2k\pi)}\right) = \log(|z|) + i(\varphi + 2k\pi)$$

```python
# Imaginary Parts

def icRoot1(r, p):
    return r**(1/2) * np.sin(p/2)

def icRoot2(r, p):
    return r**(1/2) * np.sin(p/2 + np.pi)

def icLog(r, p):
    return p

x = np.linspace(-2, 2, 50)
y = np.linspace(-2, 2, 50)

X, Y = np.meshgrid(x, y)

R = np.sqrt(X**2 + Y**2)
P = np.atan2(Y, X)

zR1 = icRoot1(R, P)
zR2 = icRoot2(R, P)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

ax.plot_surface(X, Y, zR1, linewidth=0, antialiased=True)
ax.plot_surface(X, Y, zR2, linewidth=0, antialiased=True)
```
